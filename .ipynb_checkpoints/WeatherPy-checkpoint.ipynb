```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress


from config import api_key


from citipy import citipy


lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python
lat_lngs = []
cities = []


lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    

    if city not in cities:
        cities.append(city)

#City Lenghts
len(cities)
```

```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"


query_url = f"{url}appid={api_key}&units={units}&q="
response = requests.get(query_url+"new york").json()


response
```

```python
city_two = []
cloudvol = []
dates = []
humidityvol = []
lats = []
longs = []
max_temps=[]
temps = []
wind_speeds = []
countries = []
count = 0
set_ = 1

# loops for creating columns
for city in cities:
    try:
        response = requests.get(query_url + city.replace(" ","&")).json()
        cloudvol.append(response['clouds']['all'])
        countries.append(response['sys']['country'])
        dates.append(response['dt'])
        humidityvol.append(response['main']['humidity'])
        lats.append(response['coord']['lat'])
        longs.append(response['coord']['lon'])
        temps.append(response['main']['temp'])
        max_temps.append(resonse['main']['temp_max'])
        wind_speeds.append(response['wind']['speed'])
        if count > 48:
            count = 1
            set_ += 1
            city_two.append(city)
        else:
            count += 1
            city_two.append(city)
        print(f"Processing {count} of {set_}, City Name: {city}")
    except Exception:
        print("Location not found. Going to next data")
print("\nSearch Complete")
```

```python
weather_1 = {
    "City":city_two,
    "Country":countries,
    "Date":dates,
    "Lat":lats,
    "Long":longs,
    
    "Temperature":temps,
    "Humidity":humidityvol,
    "Wind Speed":wind_speeds,
    "Cloud Cover":cloudvol
}
```

```python
x_values = weather_north['Lat']
y_values = weather_north['Temperature']
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

lat_temp_reg_n = plt.scatter(x_values,y_values,marker='o',edgecolors='blue')
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq, xy=(0,32),fontsize=15,color='red')
plt.grid()
plt.title('Temperature vs. City Latitude of Northern Hemisphere',size=18)
plt.xlabel('Latitude',size=15)
plt.ylabel('Temperature (f)',size=15)
plt.savefig('C:/Users/Brendan/Documents/GitHub/python-api-challenge/NorthernHemisphereLatitudeVsTemperature.png')
plt.show()

print(f'The r squared is: {rvalue}')
#The findings below (specifically the negative R Squared value) suggest a strong indication of tempreture dropping as cities are further North from Latitude 0. 
```
