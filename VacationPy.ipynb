```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

from config import google_key

gmaps.configure(api_key=google_key)
```

```python
csv = "C:/Users/Brendan/Documents/GitHub/python-api-challenge/weather_cities2.csv"
mappingdf = pd.read_csv(csv)
mappingdf.set_index = None
mappingdf.head()
```

```python

locations = mappingdf[["Lat", "Long"]]
humidity = mappingdf['Humidity']/100
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=5,
                                 point_radius=5)


fig.add_layer(heat_layer)

fig
```

```python
narrowed_df = mappingdf.copy()
narrowed_df = mappingdf[mappingdf['Temperature'] >= 70]
narrowed_df = mappingdf[mappingdf['Temperature'] <= 80]  
narrowed_df = mappingdf[mappingdf['Wind Speed'] <= 10]
narrowed_df = mappingdf[mappingdf['Cloud Cover'] == 0]
narrowed_df = narrowed_df.dropna()
narrowed_df = narrowed_df.reset_index()
narrowed_df
#I understand the question asked to filter by max_temp, however due to the 
#API Key limitation I could not make a new table and filtered by the averagte Temperature
```

```python
hotel_df = narrowed_df
hotel_df["Hotel Name"] = np.nan
hotel_df = hotel_df.drop(columns='index')
hotel_df
```

```python
hotels = []
x = 0

# loop to search through city coords
for cities in hotel_df['City']:
    target_coordinates = str(hotel_df['Lat'][x])+' , '+str(hotel_df['Long'][x])
    x+=1
    target_radius = 5000
    target_type = 'lodging'
    params = {
    "location": target_coordinates,
    "radius": target_radius,
    "type": target_type,
    "key": google_key
    }
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
# try/except to check if a city turns up    
    try:
        response = requests.get(base_url, params=params)
        hotel_results = response.json()
        hotels.append(hotel_results['results'][0]['name'])
        
    except IndexError:
        hotels.append(np.nan)

#append dataframe with hotel names
hotel_df['Hotel Name'] = hotels
hotel_df
```

```python
hotel_df_clean = hotel_df.dropna()
hotel_df_clean
```

```python
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
hotel_info = [info_box_template.format(**row) for index, row in hotel_df_clean.iterrows()]
locations = hotel_df_clean[["Lat", "Long"]]
```

```python
fig = gmaps.figure()
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
fig.add_layer(heat_layer)
fig
```
